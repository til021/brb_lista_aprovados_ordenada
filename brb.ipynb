```python
import tabula
import pandas as pd
```

```python
# Obtendo as notas objetivas:
objetivas = tabula.read_pdf("objetivas_brb.pdf", pages='all', lattice=True, encoding='latin-1', multiple_tables=False)
```

```python
df_obj = pd.DataFrame()

for item in objetivas:
    df_obj = pd.concat([df_obj, pd.DataFrame(item)], axis =0)
```

```python
df_obj.info()
```

```python
df_obj.head()
```

```python
# Obtendo os aprovados (DA AMPLA CONCORRÊNCIA) que tiveram a redação corrigida:
aprovados_objetiva = df_obj[df_obj['Unnamed: 15']== 'Sim']
```

```python
aprovados_objetiva = aprovados_objetiva[['Inscrição', 'Candidato', 'Unnamed: 10']]
```

```python
# Limpando os dados:
aprovados_objetiva.rename(columns={'Unnamed: 10': 'Nota_Final_Objetiva'}, inplace=True)
```

```python
aprovados_objetiva['Nota_Final_Objetiva'] = aprovados_objetiva['Nota_Final_Objetiva'].str.replace(',','.')
aprovados_objetiva['Nota_Final_Objetiva'] = aprovados_objetiva['Nota_Final_Objetiva'].astype(float)
```

```python
aprovados_objetiva['Inscrição'] = aprovados_objetiva['Inscrição'].astype(int).astype(str)
```

```python
aprovados_objetiva.reset_index(drop=True, inplace=True)
```

```python
aprovados_objetiva.index = aprovados_objetiva.index+1
```

```python
aprovados_objetiva.head()
```

```python
aprovados_objetiva.info()
```

```python
discursivas = tabula.read_pdf("discursiva_brb.pdf", pages='all', lattice=True, encoding='latin-1', multiple_tables=False)
```

```python
df_disc = pd.DataFrame()

for item in discursivas:
    df_disc = pd.concat([df_disc, pd.DataFrame(item)], axis =0)
```

```python
df_disc.info()
```

```python
df_disc.head()
```

```python
aprovados_disc = df_disc[df_disc['Unnamed: 14']== 'Aprovado']
```

```python
aprovados_disc = aprovados_disc[['Inscrição', 'Unnamed: 13']]
```

```python
# Limpando os dados:
aprovados_disc.rename(columns={'Unnamed: 13': 'Nota_Preliminar_Discursiva'}, inplace=True)
```

```python
aprovados_disc['Nota_Preliminar_Discursiva'] = aprovados_disc['Nota_Preliminar_Discursiva'].str.replace(',','.')
aprovados_disc['Nota_Preliminar_Discursiva'] = aprovados_disc['Nota_Preliminar_Discursiva'].astype(float)
```

```python
aprovados_disc['Inscrição'] = aprovados_disc['Inscrição'].astype(int).astype(str)
```

```python
aprovados_disc.head()
```

```python
aprovados_disc.info()
```

```python
resultado_preliminar = aprovados_objetiva.merge(aprovados_disc, how='inner', on='Inscrição')
```

```python
resultado_preliminar
```

```python
# Observe que temos 980 aprovados na ampla concorrência que tiveram a redação corrigida,
# Por outro lado, tivemos 895 candidatos aprovados na prova discursiva, incluindo candidatos da ampla, pnp, pcd e hipossuficientes.

# A quantidade total de aprovados na ampla foi de 738 candidatos.
```

```python
resultado_preliminar['NPF'] = resultado_preliminar['Nota_Final_Objetiva'] + resultado_preliminar['Nota_Preliminar_Discursiva']
```

```python
resultado_preliminar = resultado_preliminar.sort_values('NPF', ascending=False).reset_index(drop=True)
```

```python
resultado_preliminar.index = resultado_preliminar.index + 1
```
